```python
#24. Write a Python program to find the second most repeated word in a given string
text = "Data science is the study of data Data Data Data science"

words = text.lower().split()

count = {}
for w in words:
    count[w] = count.get(w, 0) 

sorted_words = sorted(count, key=count.get, reverse=True)

print("Second most repeated word:", sorted_words[1])
print(sorted_words)
```
